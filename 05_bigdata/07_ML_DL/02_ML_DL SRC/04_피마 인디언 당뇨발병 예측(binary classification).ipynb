```python
#csv파일을 numpy배열로 읽어오기
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
dataset.shape
```

```python
# 훈련셋 : 검증셋 = 7:3
# 학습용 데이터(훈련데이터=훈련셋)
x_train = dataset[:700, :-1]
y_train = dataset[:700, -1]

# 테스트 데이터(시험데이터 = 시험셋)
x_test = dataset[700:, :-1]
y_test = dataset[700:, -1]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
#독립면수가 8게 타겟변수가 1(원 핫인코딩을 안 함, 0, 1)
#1  --(원핫인코딩) ->01
#0  --(원핫인코딩) ->10
```

# 2. 모델 구성

```python
model = Sequential()
# input_dim은 8, 최종출력 1
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

# 3. 학습과정 설정

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 4. 학습시키기

```python
hist = model.fit(x_train, y_train, epochs=1500, batch_size=64)
```

# 5. 모델 평가하기(학습과정보기, 평가)

```python
hist.history.keys()
```

```python
# 모델의 학습과정 표시하기
fig, loss_ax = plt.subplots(figsize=(10,6))
loss_ax.plot(hist.history['loss'],'r',label='loss')
acc_ax = loss_ax.twinx() # loss_ax와 x축 공유
acc_ax.plot(hist.history['accuracy'],'b',label='accuracy')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='center left')
plt.show()
```

```python
#모델 평가하기
score = model.evaluate(x_test, y_test)
```

```python
print("loss : %s, accuracy : %s" % (score[0], score[1]))
```

# 6. 모델 사용하기

1. 데이터셋 가져와 훈련셋, 검증셋, 독립변수와 타겟변수. 타겟변수는 원핫 인코딩

4. 학습하기
5. 평가하기
6. 사용하기

```python
import numpy as np # csv 파일을 numpy 배열로 읽기
import pandas as pd
import sys
import tensorflow.keras.utils as utils # 분류분석 원핫인코딩시
from tensorflow.keras.models import Sequential # 모델 생성시
from tensorflow.keras.layers import Dense, Activation
import matplotlib.pyplot as plt
```

# 1. 데이터 셋 생성 vs 전처리

```python
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
dataset.shape
```

```python
# 훈련셋 : 검증셋 = 7:3
# 학습용 데이터(훈련데이터=훈련셋)
x_train = dataset[:700, :-1]
y_train = dataset[:700, -1]

# 테스트 데이터(시험데이터 = 시험셋)
x_test = dataset[700:, :-1]
y_test = dataset[700:, -1]
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
y_train = utils.to_categorical(y_train,2)
y_test = utils.to_categorical(y_test,2)
```

```python
y_test, y_test.shape
```

# 2. 모델 구성

```python
model = Sequential()
# input_dim은 8, 최종출력 1
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

# 3. 학습과정 설정

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

# 4. 학습시키기

```python
hist = model.fit(x_train, y_train, epochs=1500, batch_size=64)
```

# 5. 모델 평가하기(학습과정보기, 평가)

```python
# 모델의 학습과정 표시하기
fig, loss_ax = plt.subplots(figsize=(10,6))
loss_ax.plot(hist.history['loss'],'r',label='loss')
acc_ax = loss_ax.twinx() # loss_ax와 x축 공유
acc_ax.plot(hist.history['accuracy'],'b',label='accuracy')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='center left')
plt.show()
```

```python
#모델 평가하기
score = model.evaluate(x_test, y_test)
```

```python
print("loss : %s, accuracy : %s" % (score[0], score[1]))
```
