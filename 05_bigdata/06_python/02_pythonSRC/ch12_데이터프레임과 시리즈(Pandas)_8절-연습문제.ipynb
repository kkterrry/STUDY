## 11.1 fillna (결측치를 어떤 값으로 바꿈)

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3,4,np.nan,1],
                   [np.nan, 3, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                   columns=list('ABCD'))
df
```

```python
# 모든 결측치를 무조건 0으로 바꿈
df.fillna(0)
```

```python
# 결측치를 이전 값이나 다음 값으로 채움
        #method = pad/ffill : 결측치가 이전 값으로 대체
        #method = backfill/bfill : 결측치를 다음 값으로 대체
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
# 결측치 특정대체값으로 바꿈
values = {'A':99, 'B':50, 'C':0, 'D':95} # 열별 결측치 대체값
df.fillna(value=values)
```

```python
df.fillna(value = df.mean()) # 모두 결측치인 c열은 대체가 안된다.
```

```python
df.fillna(value=values, limit=3) # 3번째 결측치까지만 대체
```

## 11.2 replace(to_value, new_value, inplace=True)
- to_value를 new_value로 변경
- inplace=False : 변경된 내용을 반환. 데이터 프레임에는 적용되지 않음
- inplace=True : 변경된 내용을 반환하지 않고. 데이터 프레임에는 적용
- 정규표현식을 이용 가능

```python
s = pd.Series([0,1,2,3,4,None])
s.replace(np.nan, 0, inplace=True)
s.replace([0,1],5) # 0과1을 5로 변경
```

```python
s = pd.Series([0,1,2,3,4,None])
s.replace(range(0,3),range(100,103))
```

```python
df = pd.DataFrame({'A':['bat','foo','bait'],
                   'B':['abc','bar','xyz']})
df
```

```python
# to_value에 값, 정규표현식(raex=True인자 추가)
df.replace({'A':'bat','B':'xyz'},'XXX')
```

```python
# to_value에 값, 정규표현식(raex=True인자 추가)
df.replace(r'^ba.+$','NEW',regex=True)
```

## 11.3 where(조건), mask(조건)
* where(조건) : 조건이 만족하는 요소는 그대로 출력. (반)mask

```python
s = pd.Series(range(10,15))
s
```

```python
s.where(s>11) #조건에 만족하는 값만 보이고 조건에 만족하지 않으면 보이지 않음
```

```python
s.where(s>11, -99)  #조건에 만족하면 보이고 만족하지 않으면 -99
```

```python
s.mask(s>11) #조건에 만족하는 경우만 결측치. 만족하지 않으면 값 보임
```

```python
s.mask(s>11,-s) #조건에 만족하는 경우 -s값으로 출력
```
