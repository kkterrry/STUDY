### 2) 네이버 영화 랭킹 출력하기

```
9위 영화제목 - 영화소개 url
감독 및 배우
아무개: 감독 / 저무개 : 주연 / 이무개 : 주연
```

````python
import requests
from bs4 import BeautifulSoup
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
movie_ranking = requests.get(url)
soup = BeautifulSoup(movie_ranking.content,"html.parser")
title_list = soup.select('td.title > div.tit3 > a')
for idx in range(len(title_list)):
    link = url + title_list[idx].attrs['href']
    print("{:2d}위 영화 제목 : {} - {}".format(idx+1, title_list[idx].text, link))
    
    page = requests.get(link)
    s = BeautifulSoup(page.content,"html.parser")
    people = s.select('li > a.tx_people')  #사람이름
    staffs = s.select('dl.staff > dt')  #역할
    print("감독 및 배우")
    for i in range(len(people)):
        print("{} : {}".format(people[i].text, staffs[i].text), end="  /  ")
    print('\n')
````

````python
import requests
from bs4 import BeautifulSoup
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
movie_ranking = requests.get(url)
soup = BeautifulSoup(movie_ranking.content,"html.parser")
title_list = soup.select('td.title > div.tit3 > a')
dic_list = []

for idx in range(len(title_list)):
    dic = {} #빈딕셔너리
    dic['rank'] = idx+1
    dic['title'] = title_list[idx].text
    dic['link'] = url + title_list[idx].attrs['href']
    
    page = requests.get(dic['link'])
    s = BeautifulSoup(page.content,"html.parser")
    people = s.select('li > a.tx_people')  #사람이름
    staffs = s.select('dl.staff > dt')  #역할
    
    dic['staff'] = ''
    for i in range(len(people)):
        dic['staff'] += "{}:{} /".format(people[i].text, staffs[i].text)
    dic_list.append(dic)

import pandas as pd
df = pd.DataFrame(dic_list)
df.to_csv('ch14_movie.csv', encoding='cp949')
df
````

    * https://goo.gl/HLntbd
    * https://www.nextree.co.kr/p4327
    * 정규표현식 연습장 : https://regexr.com
간략 문법:
    \d : 숫자와 매치, [0-9]와 동일
    \D : 숫자가 아닌 것
    \s : whitespace
    \w : 영문자나 숫자
    \W : 영문자나 숫자가 아닌 문자
    .  : 문자하나
    +  : 1번 이상 반복
    *  : 0번 이상 반복
    {2,4} : 2~4회 반복
    ? : {0,1}의 의미
    ^ : 시작
    $ : 끝
정규표현식의 예 ; 메일 정규표현식, 전화번호 정규표현식, ip주소 정규표현식, 주민번호 정규표현식

### 전국 날씨 Rss데이터 가져오기
구글에서 전국날씨RSS 검색후 나온 결과 사이트를 방문 -> 전국RSS열기(링크 들오면된다)

````python
import re, urllib.request
url = "http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"
target = urllib.request.urlopen(url)
contents = target.read().decode('utf-8') #한글이 있는 경우
# re.DOTALL: 정규표현식의 .에 줄바꿈문자(\n)를 포함
locations = re.findall(r'<location wl_ver="3">.+?</location>', contents, re.DOTALL)
for loc in locations:
    province = re.search(r'<province>(.+)</province>',loc)
    city = re.search(r'<city>(.+)</city>',loc)
    print(province.group(1),"지역의 ",city.group(1), '도시의 날씨입니다')
    data = re.findall(r'<data>.+?</data>',loc, re.DOTALL) #\n과 같은것도 .에 포함된다.
    ampm = '오전'
    for item in data:
        tmEf = re.search(r'<tmEf>(.+)</tmEf>',item).group(1)  #날짜 및 시간
        wf = re.search(r'<wf>(.+)</wf>',item).group(1) # 날씨
        tmn = re.search(r'<tmn>(.+)</tmn>',item).group(1) #최저기온
        tmx = re.search(r'<tmx>(.+)</tmx>',item).group(1) #최저기온
        print('\t날짜시간:{} {}, 날씨:{}, 최저:{}도, 최고:{}도'.format(tmEf[:10], ampm, wf, tmn, tmx))
        if ampm == '오전':
            ampm ='오후'
        else:
            ampm ='오전'
````

````python
# BeautifulSoup을 이용해 xml 크롤링
url = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
# target = request.get(url)
# soup = BeautifulSoup(target.content, "xml")
target = urlopen(url)
soup = BeautifulSoup(target, "xml")
locations = soup.find_all("location")
#locations = soup.select("location")
for loc in locations:
    province = loc.find('province').text
    #province = loc.select_one('province').text
    city = loc.find('city').text
    #city = loc.select_one('city').text
    data = loc.find_all('data')
    # data = loc.select('data')
    print(province,"지역의",city,"도시 날씨입니다.")
    ampm = '오전'
    
    for item in data:
        tmEf = item.find('tmEf').text[:10]
        wf = item.find('wf').text
        tmn = item.find('tmn').text
        tmx = item.find('tmx').text
        print("\t{} {} 날씨:{}, 최저:{}도, 최고:{}도".format(tmEf, ampm, wf, tmn, tmx))
        if ampm == '오전':
            ampm ='오후'
        else:
            ampm ='오전'
````
